```python
"""Activation functions introduce non-linearity in data and help in better training.
Here, we shall explore the different activation funcions commonly used."""

#sigmoid function = 1/(1+exp(-x))
import matplotlib.pyplot as plt
import numpy as np

n = np.linspace(-1, 1)
#plt.plot(n, n, label='$n$')
sig = 1/(1+np.exp(-n))
plt.plot(n, np.exp(n), label='exp')
plt.plot(n, sig, label='sigmoid function')
```
