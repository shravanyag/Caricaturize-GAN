```python
"""Plotting a point on linear vs logarithmic scale"""
import numpy as np
import matplotlib.pyplot as plt

# Probability value
Q_x = 0.6

# Calculate the logarithm of Q(x)
log_Q_x = np.log(Q_x)

# X-axis values
x = np.linspace(0.001, 0.999, 100)

# Y-axis values for Q(x)
y_Q_x = [Q_x] * len(x)

# Y-axis values for log(Q(x))
y_log_Q_x = [log_Q_x] * len(x)

# Plot Q(x) and log(Q(x))
plt.plot(x, y_Q_x, label='Q(x)')
plt.plot(x, y_log_Q_x, label='log(Q(x))')
plt.xlabel('x')
plt.ylabel('Probability')
plt.title('Q(x) and log(Q(x))')
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""Understanding and plotting loss functions in ML models.
In this code, we will be plotting the binary cross entropy loss function.
Instead of taking two distributions, we will compare the distribution with itself."""

import matplotlib.pyplot as plt
import numpy as np

n = np.linspace(0.1, 10)
plt.plot(n, n, label='$n$')
plt.plot(n, (np.log(n)), label='$(log n)$')
plt.plot(n, -1*n*np.log(n), label='loss function')
plt.legend()
```

```python
#Above we observe that the logarithmic scale shows variations between distributions in a pronounced way
#Thus making it a better scale to represent our loss function on
```
